Assume that the distance between microphones is $30 cm = 0.3m$. This makes $x_b = 0.15$.

Assume that $ c = 343 m/s $

We need to first find $\tau$, and then $AB' = \tau \cdot c$

$$ y = \sqrt{ \frac{AB'^2}{4} - x_B^2 + x^2(\frac{4x_b^2}{AB'^2} - 1)} $$

Since $$ AB \leq 2 * x_B$$
We get $$ \tau \leq \frac{2*x_B}{c} = \frac{2*0.15}{343} = 874\mu s$$ 



```python
def y(AB, x, xb = 0.15):
    try:
        yy = np.sqrt(AB**2/4 - xb**2 + x**2 * (4*xb**2/AB**2 - 1))
        return yy
    except Exception:
        return np.nan

def negy(AB, x, xb = 0.15):
    return -y(AB, xb, x)


c = 343


```

## Test 2
The original beep starts at 0.80870 seconds. 

Let's delay it by 850 us to start at 0.80955 seconds.

```python
def anglefromfiles(delayA, delayB, filename="./sounds/Beep_delay_"):
    # put these names in
    # depending on which mic you want to test
    filenameA = filename + str(delayA) + "us.wav"
    filenameB = filename + str(delayB) + "us.wav"

    (rateA, dataA) = wavfile.read(filenameA)
    (rateB, dataB) = wavfile.read(filenameB)

    # find first point where the beep starts
    # primitive method to find tau - basic threshold based system
    locA = np.argmax((dataA > 100 )|( dataA < -100))
    locB = np.argmax((dataB > 100 )|( dataB < -100))

    tau = (locB - locA)/rateA
    if tau == 0:
        return 0
    print("tau=", tau)
    AB = tau*c
    print("AB=", AB)

    xxx = np.linspace(-5, 5, 10000)
    yyy = y(AB, xxx)

    plt.plot(xxx, yyy)
    plt.show()

    # The approximate angle is -12 degrees from straight
    degs = np.degrees(np.arctan(np.gradient(yyy, xxx)))

    # sound hits B after A, left quadrant
    if tau > 0:
        angle = -(90 - degs[-1])
    else:
        # right quadrant
        angle = 90 - degs[-1]


    print(angle, " degrees from center")
    return angle
```

```python
delays = [850, 800, 700, 600, 500, 400, 300, 200, 100, 50, 0, -50, -100, -200, -300, -400, -500, -600, -700, -800, -850]
angles = []
for delay in delays:
    if delay < 0:
        angles.append(anglefromfiles(delayA=-delay, delayB=0))
    else:
        angles.append(anglefromfiles(delayA=0, delayB=delay))

```

```python
plt.plot(delays, angles)
plt.xlabel("Delay (B-A) in us")
plt.ylabel("Angle from Center")
plt.grid()
plt.title("Angle vs Delay")
```

```python
import statsmodels.api as sm
from numpy.lib.stride_tricks import as_strided


def _check_arg(x, xname):
    x = np.asarray(x)
    if x.ndim != 1:
        raise ValueError('%s must be one-dimensional.' % xname)
    return x


def crosscorrelation(x, y, maxlag):
    """
    Cross correlation with a maximum number of lags.

    `x` and `y` must be one-dimensional numpy arrays with the same length.

    This computes the same result as
        numpy.correlate(x, y, mode='full')[len(a)-maxlag-1:len(a)+maxlag]

    The return vaue has length 2*maxlag + 1.
    https://stackoverflow.com/questions/30677241/how-to-limit-cross-correlation-window-width-in-numpy
    """
    x = _check_arg(x, 'x')
    y = _check_arg(y, 'y')
    py = np.pad(y.conj(), 2*maxlag, mode='constant')
    T = as_strided(py[2*maxlag:], shape=(2*maxlag+1, len(y) + 2*maxlag),
                   strides=(-py.strides[0], py.strides[0]))
    px = np.pad(x, maxlag, mode='constant')
    return T.dot(px)


def findLocAutoCorrelation(rateA, dataA, rateB, dataB):
    # we want to calculate how many samples off they are
    # but only up till 850us in sample
    
    # 850us * 44.1samples/sec
    
    
    dataA = dataA.flatten()
    dataB = dataB.flatten()
    max_sample_diff = (int)(np.ceil(850e-6 * rateA))
    
    if len(dataA) < len(dataB):
        dataA = np.pad(dataA, (len(dataB) - len(dataA), 0), 'constant', constant_values=(0))
    elif len(dataA) > len(dataB):
        dataB = np.pad(dataB, (len(dataA) - len(dataB), 0), 'constant', constant_values=(0))
    
    """
    lagsB = np.zeros((max_sample_diff))
    
    for i in range(max_sample_diff):
        lagsB[i] = np.corrcoef(dataA, dataB[i:])
    
    lagsA = np.arange(max_sample_diff)
    for i in range(max_sample_diff):
        lagsA[i] = np.corrcoef(dataA[i:], dataB)

    lag0 = np.corrcoef(dataA, dataB)
    """
        
    xcc = crosscorrelation(dataA, dataB, max_sample_diff)
    
    # take where it is max, convert it back to sample and calculate delay:
    tau = (np.argmax(xcc) - max_sample_diff)/rateA
    print("sample diff:", (np.argmax(xcc) - max_sample_diff))
    print("tau: ", tau)
    return tau
    
    
```
